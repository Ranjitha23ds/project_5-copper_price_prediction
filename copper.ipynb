```python
Q1 = df['quantity tons'].quantile(0.25)
Q3 = df['quantity tons'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['quantity tons'] < lower_limit) | (df['quantity tons'] > upper_limit)]
outliers.shape
```

```python
# for quantity tons i use capping for outlier treatment
#some customers buy small samples, others bulk.
#Some values are very small (e.g., 0.09, 0.13, etc.)

#Resulting price_per_ton becomes very high (e.g., 13000, 7000, etc.)
```

```python
#capping for quatity tons
import numpy as np
df["quantity tons"]=np.where(df["quantity tons"]>upper_limit,upper_limit,
                    np.where(df["quantity tons"]<lower_limit,lower_limit,df["quantity tons"]))
```
